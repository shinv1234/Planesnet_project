```python
%%time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
resnet_train_acc_list = np.loadtxt('resnet_train_acc_list.txt')
resnet_test_acc_list = np.loadtxt('resnet_test_acc_list.txt')
resnet_cost_list = np.loadtxt('resnet_cost_list.txt')
plain_model_train_acc_list = np.loadtxt('plain_model_train_acc_list.txt')
plain_model_test_acc_list = np.loadtxt('plain_model_test_acc_list.txt')
plain_model_cost_list = np.loadtxt('plain_model_cost_list.txt')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
ax[0].plot(plain_model_train_acc_list)
ax[0].plot(plain_model_test_acc_list)
fig.suptitle("Plain_Model", fontsize=18, ha='center')
ax[0].grid(True)
ax[0].set_title("Plain_Model's Train & Test Accuracy")
ax[0].set_xlabel("Epoch")
ax[0].set_ylabel("Accuracy")
ax[0].legend(['Train Accuracy', 'Test Accuracy'])
ax[1].plot(plain_model_cost_list)
ax[1].grid(True)
ax[1].set_title("Plain_Model's Cost")
ax[1].set_xlabel("Epoch")
ax[1].set_ylabel("Cost")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
ax[0].plot(resnet_train_acc_list)
ax[0].plot(resnet_test_acc_list)
fig.suptitle("ResNet", fontsize=18, ha='center')
ax[0].grid(True)
ax[0].set_title("ResNet's Train & Test Accuracy")
ax[0].set_xlabel("Epoch")
ax[0].set_ylabel("Accuracy")
ax[0].legend(['Train Accuracy', 'Test Accuracy'])
ax[1].plot(resnet_cost_list)
ax[1].grid(True)
ax[1].set_title("ResNet's Cost")
ax[1].set_xlabel("Epoch")
ax[1].set_ylabel("Cost")
plt.show()
```
